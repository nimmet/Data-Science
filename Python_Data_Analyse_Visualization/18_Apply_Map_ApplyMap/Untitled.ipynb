```python
def get_age_group(age):
    if age < 2:
        return "infant"
    elif age < 12:
        return "child"
    elif age < 18:
        return "teen"
    elif age <50:
        return "adult"
    else:
        return "senior"

    
titanic.age.apply(get_age_group)
```

```python
def get_age_mean(age):
    if age.mean()<12:
        return "child"
    elif age.mean()<18:
        return "teen"
    elif age.mean()<50:
        return "adult"
    else:
        return "senior"
    
titanic.age.apply(get_age_mean)
```
