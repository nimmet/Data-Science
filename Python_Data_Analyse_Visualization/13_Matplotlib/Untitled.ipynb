# Bar Plots

```python
plants = ["Spinach","Turip","Rhubarb","Broccoli","Kale"]
died = [10,25,5,30,21]
d1= [i+2 for i in died]
d2= [i+2 for i in d1]
germinated = [74,88,56,69,59]
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, germinated, width=0.6, color="orchid")
plt.title("Germination Rates")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, width=0.6, color="olive")
plt.title("Died Rates")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="olive", width=0.5,align="edge")
plt.bar(plants, germinated, width=0.3, color="orchid")
plt.title("Common Rates")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="olive", width=0.3,align="edge", label="Died")
plt.bar(plants, germinated, width=0.3, color="orchid",align="center", label="Germinated")
plt.bar(plants, d1, color="green", width=-0.3,align="edge", label="Lived")
plt.bar(plants, d2, color="red", width=0.3, label="Unamed")
plt.title("Common Rates")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, color="olive", width=0.5, bottom=germinated, label="Failed Seedlings")
plt.bar(plants, germinated, width=0.5, color="orchid", label="Germinated Seedlings")
plt.title("Common Rates")
plt.legend(shadow=True,frameon=True,facecolor="white")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.barh(plants, died, color="olive")
plt.show()
```
